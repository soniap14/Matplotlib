```python
Pyber Ride Sharing Analysis
• The Bubble Plot indicates that number of rides and drivers are more in urban cities as compared to suburban cities and rural cities. Interestingly average fare is way more in suburban cities compared to urban and rural cities.

• All three pie charts indicate that urban cities have a considerably high percentage of total fare, rides and drivers compared to suburban and rural cities.

• From this analysis, one can predict that there are new opportunities to expand in suburban cities. There is ride demand but there are very few Drivers
```

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

#City data
city_df.head()
```

```python
#Drop Duplcates

city_df= city_df.drop_duplicates('city')
city_df.head()
```

```python
#ride data
ride_df.head()
```

```python
# combine both data sets# combi 
comabine_df = city_df.merge(ride_df, on = 'city', how = 'outer')
comabine_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#city types - Urban, Suburban and Rural
urban = combine_df.loc[(combine_df["type"] == "Urban")]
suburban = combine_df.loc[(combine_df["type"] == "Suburban")]
rural= combine_df.loc[(combine_df["type"] == "Rural")]
```

```python
# Per city cals of Average Fare, Total Rides and Total Drivers for Urban cities
avg_fare_urban = urban.groupby(['city'])['fare'].mean()
total_rides_urban = urban.groupby(['city']).count()['ride_id']
total_drivers_urban = urban.groupby(['city'])['driver_count'].value_counts()

# Per city cals of Average Fare, Total Rides and Total Drivers for Suburban cities
avg_fare_suburban = suburban.groupby(['city'])['fare'].mean()
total_rides_suburban = suburban.groupby(['city']).count()['ride_id']
total_drivers_suburban = suburban.groupby(['city'])['driver_count'].value_counts()

# Per city cals of Average Fare, Total Rides and Total Drivers for Rural cities
avg_fare_rural = rural.groupby(['city'])['fare'].mean()
total_rides_rural = rural.groupby(['city']).count()['ride_id']
total_drivers_rural = rural.groupby(['city'])['driver_count'].value_counts()
```

```python
# Build the scatter plots for each city types # Incorporate the other graph properties
plt.scatter(total_rides_urban, avg_fare_urban,s=total_drivers_urban*10,
            marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.75, label="Urban")

plt.scatter(total_rides_suburban, avg_fare_suburban,s=total_drivers_suburban*10,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.75, label="Suburban")

plt.scatter(total_rides_rural, avg_fare_rural,s=total_drivers_rural*10,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.75, label="Rural")

plt.grid()


# Chart title
plt.title("Pyber Ride Sharing Data")
# x label
plt.xlabel("Total Numbers of Rides (Per City))")
# y label
plt.ylabel("Average Fare($)")

# Create a legend
plt.legend(loc='upper right')

# Save Figure
plt.savefig("../Images/Pyber Ride Share Bubble plot.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Total Fares by City Type in Percent

# Calculate Total Fare by City 
total_fare = combine_df.groupby(['type'])['fare'].sum()
# Labels for the sections of our pie chart
labels = ["Rural","Suburban","Urban" ]
# The colors of each section of the pie chart
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)

# Build Pie Chart
plt.title("% of Total Fares By City Types")
plt.pie(total_fare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)

# Save Figure
plt.savefig("../Images/Total Fares By City pie chart.png")

# Show plot
plt.show()
```

## Total Rides by City Type

```python
# Calculate Total Rides by City Type in Percents

# Calculate Total Fare by City Type
total_rides = combine_df.groupby(['type'])['ride_id'].count()
# Labels for the sections of our pie chart
labels = ["Rural","Suburban","Urban" ]
# The colors of each section of the pie chart
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)

# Build Pie Chart
plt.title("% of Total Rides By City Types")
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

# Save Figure
plt.savefig("../Images/Total Rides By City pie chart.png")

# Show plot
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Total Drivers by City Type in Percents

# Calculate Total Drivers by City Type# Calcul 
total_drivers = city_df.groupby(['type'])['driver_count'].sum()
# Labels for the sections of our pie chart
labels = ["Rural","Suburban","Urban" ]
# The colors of each section of the pie chart
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)


# Build Pie Charts
plt.title("% of Total Drivers By City Types")
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

# Save Figure
plt.savefig("../Images/Total Drivers By City pie chart.png")

# Show plot
plt.show()
```
